```python
# This tells matplotlib not to try opening a new window for each plot.
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
import numpy as np
import pandas as pd
from sklearn.cross_validation import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
# dataset is by driver by day
df_trip = pd.read_csv('trip_forecast_data.csv')
print(df_trip.shape)
df_trip.head()
```

```python
#missing data
total = df_trip.isnull().sum().sort_values(ascending=False)
percent = (df_trip.isnull().sum()/df_trip.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(10)
```

```python
# select features and remove null for key variables
columns = ['tripstartmonth', 'weekday','pickupcommunity', 'prcp', 'snow', 'snwd', 'tmax', 'tmin',
       'tobs', 'tripcount', 'totalfare','duration']
df1 = pd.DataFrame(df_trip, columns=columns)
df1 = df1.dropna()
df1['pickupcommunity']=df1['pickupcommunity'].astype(int).astype(str)
df1['tripstartmonth']=df1['tripstartmonth'].astype(str)
top_community = ['8','32','28','76','6','7','33','24','56','3']
df1['pickupcommunity']=[c if c in top_community else 'other' for c in df1['pickupcommunity']]
# # replace missing value with 0 for pickup community
# df1['pickupcommunity'] = df1['pickupcommunity'].fillna(0)
# df1['duration'] = df1['duration'].fillna(0)

# remove missing value

df1.describe()
```

```python
# treat month/day/weekday as categorical variables
categorical = ['tripstartmonth', 'weekday','pickupcommunity']
numerical = ['prcp', 'snow', 'snwd', 'tmax', 'tmin','tobs']
# create dummy variables for categorical features
df1_dummy = pd.get_dummies(df1[categorical])
df1_dummy.head()
```

```python
# combine numerical features and categorical features
df2 = pd.concat([df1[numerical], df1_dummy], axis=1)
print(df2.shape)
df2.head()
```

```python
# Partition the dataset in train + validation sets
# could choose to forecast # of trips or total fare or total duration per day per driver
y_du = df1['duration'] #'tripcount', 'totalfare','duration'
X_train, X_test, y_du_train, y_du_test = train_test_split(df2, y_du, test_size = 0.2, random_state = 0)
print("X_train : " + str(X_train.shape))
print("X_test : " + str(X_test.shape))
print("y_train : " + str(y_du_train.shape))
print("y_test : " + str(y_du_test.shape))
```

```python
y_fa = df1['totalfare'] #'tripcount', 'totalfare','duration'
X_train, X_test, y_fa_train, y_fa_test = train_test_split(df2, y_fa, test_size = 0.2, random_state = 0)
```

```python
# fit a regularized linear regression model
rdg = Ridge(alpha=4.0)  # alpha is regularization strength
rdg.fit(X_train, y_du_train)
test_pred_du = rdg.predict(X_test)
train_pred_du = rdg.predict(X_train)

print('MSE train: %.3f, test: %.3f' %
     (mean_squared_error(y_du_train, train_pred_du), mean_squared_error(y_du_test, test_pred_du)))
print('R^2 train: %.3f, test: %.3f' % 
     (r2_score(y_du_train, train_pred_du), r2_score(y_du_test, test_pred_du)))

# residual plots
plt.figure(figsize=(15, 6))
plt.subplot(1,2,1)
plt.scatter(train_pred_du, train_pred_du - y_du_train, c='blue', marker='o', label='Training data')
plt.scatter(test_pred_du, test_pred_du - y_du_test, c='green', marker='s', label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc = 'upper right')

plt.subplot(1,2,2)
plt.scatter(test_pred_du, y_du_test, c='green', marker='s')
plt.xlabel('Predicted values')
plt.ylabel('Actual values')
plt.show()
```

```python
rdg2 = Ridge(alpha=6.0)  # alpha is regularization strength
rdg2.fit(X_train, y_fa_train)
test_pred_fa = rdg2.predict(X_test)
train_pred_fa = rdg2.predict(X_train)

print('MSE train: %.3f, test: %.3f' %
     (mean_squared_error(y_fa_train, train_pred_fa), mean_squared_error(y_fa_test, test_pred_fa)))
print('R^2 train: %.3f, test: %.3f' % 
     (r2_score(y_fa_train, train_pred_fa), r2_score(y_fa_test, test_pred_fa)))
```

```python
other = ['tripstartdate','startdate_str']
df_new_dummy = pd.get_dummies(df_new[categorical])
df2_new = pd.concat([df_new[numerical], df_new_dummy], axis=1)
df2_new.head()
```
